```python
import ROOT
```

```python
_filepath = "/nfs/dust/cms/user/mykytaua/dataDeepTau/DeepTauTraining/ShuffleMergeSpectral_TrainingSpectrum/ShuffleMergeSpectral_trainingSamples-2_spectrum.root"
```

```python
file = ROOT.TFile.Open(_filepath, 'read')

th2_tau = file.Get('eta_pt_hist_tau')
th2_mu  = file.Get('eta_pt_hist_mu')
th2_jet = file.Get('eta_pt_hist_jet')
th2_el  = file.Get('eta_pt_hist_e')
```

```python
pt_tau = th2_tau.ProjectionY()
pt_mu  = th2_mu.ProjectionY()
pt_jet = th2_jet.ProjectionY()
pt_el  = th2_el.ProjectionY()

eta_tau = th2_tau.ProjectionX()
eta_mu  = th2_mu.ProjectionX()
eta_jet = th2_jet.ProjectionX()
eta_el  = th2_el.ProjectionX()
```

```python
MergeBin = 40
pt_tau.Rebin(MergeBin)
pt_mu.Rebin(MergeBin)
pt_jet.Rebin(MergeBin)
pt_el.Rebin(MergeBin)
```

```python
cv1 = ROOT.TCanvas("cv1","cv1",1000,1000)
cv1.SetLogy()
# cv1.SetLogx()

pt_tau.SetLineColor(3)
pt_mu.SetLineColor(4)
pt_jet.SetLineColor(5)
pt_el.SetLineColor(6)
pt_tau.SetLineWidth(2)
pt_mu.SetLineWidth(2)
pt_jet.SetLineWidth(2)
pt_el.SetLineWidth(2)

pt_mu.Draw("histo")
pt_tau.Draw("histosame")
# pt_mu.Draw("histosame")
pt_jet.Draw("histosame")
pt_el.Draw("histosame")

legend = ROOT.TLegend(0.4,0.75,0.7,0.94);
legend.AddEntry(pt_tau,"tau","f");
legend.AddEntry(pt_mu,"mu","f");
legend.AddEntry(pt_jet,"jet","f");
legend.AddEntry(pt_el,"e","f");
legend.Draw();

cv1.Draw()
```

```python
MergeBin = 20
eta_tau.Rebin(MergeBin)
eta_mu.Rebin(MergeBin)
eta_jet.Rebin(MergeBin)
eta_el.Rebin(MergeBin)
```

```python
cv2 = ROOT.TCanvas("cv1","cv1",1000,1000)
cv2.SetLogy()
# cv1.SetLogx()

eta_tau.SetLineColor(3)
eta_mu.SetLineColor(4)
eta_jet.SetLineColor(5)
eta_el.SetLineColor(6)
eta_tau.SetLineWidth(2)
eta_mu.SetLineWidth(2)
eta_jet.SetLineWidth(2)
eta_el.SetLineWidth(2)

eta_mu.Draw("histo")
eta_tau.Draw("histosame")
# pt_mu.Draw("histosame")
eta_jet.Draw("histosame")
eta_el.Draw("histosame")

legend = ROOT.TLegend(0.4,0.82,0.7,0.94);
legend.AddEntry(pt_tau,"tau","f");
legend.AddEntry(pt_mu,"mu","f");
legend.AddEntry(pt_jet,"jet","f");
legend.AddEntry(pt_el,"e","f");
legend.Draw();

cv2.Draw()
```
